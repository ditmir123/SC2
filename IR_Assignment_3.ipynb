```python
import numpy as np
import pandas as pd
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import ParameterEstimator, BayesianEstimator, MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
```

```python
data = pd.read_csv('heart_cleveland_upload.csv')
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data.describe()
```

```python
model = BayesianNetwork([('age','condition'),('sex','condition'),('cp','condition'),('trestbps','condition'),('chol','condition')])
```

```python
model.fit(data, estimator=MaximumLikelihoodEstimator)
```

```python
heart = VariableElimination(model)
```

```python
q1 = heart.query(variables=['condition'], evidence={'cp':3})
print(q1)
```

```python
q2 = heart.query(variables=['condition'],evidence={'trestbps':140})
print(q2)
```

```python
q3 = heart.query(variables=['condition'],evidence={'age':50})
print(q3)
```

```python
q4 = heart.query(variables=['condition'],evidence={'chol':220})
print(q4)
```
